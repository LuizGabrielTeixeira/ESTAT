```python
# ex-1 calculo de valor de parcela de emprestimo
emprestimo = float(input('Valor do empréstimo: '))
salario = float(input('Salario do comprador: '))
anos = float(input('Quantidade de anos para pagar: '))

parcela = emprestimo / (anos * 12)
valorMax = salario * 0.3

if parcela > valorMax:
    print(f'Empréstimo negado. A parcela de R${parcela:.2f} excede 30% do salário ({valorMax:.2f}).')
else:
    print(f'Empréstimo aprovado. A parcela mensal será de {parcela:.2f} e a duração sera de {anos:.0f} anos.')
```

```python
# ex-2 calculo de valor final de compra
valorCompra = float(input('Valor da compra: '))
parcelas = int(input('Quantidade de parcelas: '))

if parcelas == 1:
    valorFinal = valorCompra * 0.9
    print(f'valor sera de: {valorFinal} com opção de pagamento em {parcelas} parcelas')
elif parcelas == 2:
    valorFinal = valorCompra
    print(f'valor sera de: {valorFinal} com opção de pagamento em {parcelas} parcelas')
else:
    valorFinal = valorCompra * 1.20
    print(f'valor sera de: {valorFinal} com opção de pagamento em {parcelas} parcelas')
```

```python
# ex-3 pedra papel e tesoura
# 1 pedra
# 2 papel
# 3 tesoura
vitorias = 0
for n in range(10):
    escolha = int(input('Escolha: 1 para pedra, 2 para papel, 3 para tesoura: '))
    pcRandom = np.random.randint(1,4)
    if escolha == 1 and pcRandom == 3:
        print('Voce ganhou! escolheu pedra e o pc escolheu tesoura')
        vitorias += 1
    elif escolha == 2 and pcRandom == 1:
        print('Voce ganhou! escolheu papel e o pc escolheu pedra')
        vitorias += 1
    elif escolha == 3 and pcRandom == 2:
        print('Voce ganhou! escolheu tesoura e o pc escolheu papel')
        vitorias += 1
    else:
        print('Voce perdeu!')
print(f'Voce ganhou {vitorias} vezes! taxa de vitoria de {vitorias/10:.2f}%')
```

```python
# ex-4 lê valores positivos, sem repetir, até digitar um negativo
valores = []

while True:
    valor = float(input('Insira valores positivos ou um negativo para parar: '))
    if valor < 0:
        break

    if valor in valores:
        print('Valor já inserido, não será adicionado de novo.')
    else:
        valores.append(valor)

# se a lista não estiver vazia, mostra os resultados
if valores:
    qtd = len(valores)
    menor = min(valores)
    maior = max(valores)
    pos_menor = valores.index(menor)
    pos_maior = valores.index(maior)

    print(f'Foram inseridos {qtd} valores não repetidos.')
    print(f'O menor valor é {menor:.2f} na posição {pos_menor}.')
    print(f'O maior valor é {maior:.2f} na posição {pos_maior}.')
else:
    print('Nenhum valor válido foi inserido.')
```

```python
# ex-6
lista_par = []
lista_impar = []

for i in range(10):
    valor = int(input(f'Digite o {i+1}º valor inteiro: '))
    if valor % 2 == 0:
        lista_par.append(valor)
    else:
        lista_impar.append(valor)

lista_par.sort(reverse=True)
lista_impar.sort(reverse=True)

print(f'Valores pares em ordem decrescente: {lista_par}')
print(f'Valores ímpares em ordem decrescente: {lista_impar}')
```

```python
# ex-7
import numpy as np

dim = int(input('Input the matrix dimension: '))

matrix = np.zeros((dim, dim), dtype=int)

for i in range(dim):
    for j in range(dim):
        matrix[i,j] = int(input(f'Input the value for position [{i}][{j}]: '))

print(matrix)
print(f'Sum of even values: {matrix[matrix % 2 == 0].sum()}')
print(f'Sum of values in the third column: {matrix[:, 2].sum() if dim > 2 else 999999}')
print(f'Minor value in the second row: {matrix[1].min() if dim > 1 else 999999}')
```

```python
# ex-8

# list of students
students = []

while True:
    # list with number, name, grades and median of each student
    student = []
    studentNumber = int(input('Input the student number: '))
    studentName = input('Input the student name: ')

    # grades 1, 2 and median
    grades = []
    median = 0

    # loop to input the 2 grades
    for i in range(2):
        grade = float(input(f'Input the grade {i+1} for {studentName}: '))
        median += grade
        grades.append(grade)
    median /= 2

    # add median to the grades list
    grades.append(median)

    # add student info to the main list
    student.append(studentNumber)
    student.append(studentName)
    student.append(grades)

    students.append(student)

    # ask if want to add more students
    stop = input('More students? (y/n): ')
    if stop == 'n':
        break

# print the list of students with their info
for s in students:
    print(f'Number: {s[0]}, Name: {s[1]}, Grades: {s[2][0]:.2f}, {s[2][1]:.2f}, Median: {s[2][2]:.2f}')

```

```python
# ex-9
def counter(start, end, step):
    for i in range(1,11,1):
        print(i)
    for i in range(10,0,-2):
        print(i)
    for i in range(start, end+1, step):
        print(i)

start = int(input('Input the start number: '))
end = int(input('Input the end number: '))
step = int(input('Input the step: '))

counter(start, end, step)
```

```python
# ex-10
import numpy as np

def greater(values):
    greater = float('-inf')
    for i in range(len(values)):
        if values[i] > greater:
            greater = values[i]
    return greater

values = np.random.randint(0, 100, size=15, dtype=int)
print(values)
print(greater(values))
```

```python
# ex 10.2 using ((* values))

import numpy as np

def greater(* values):
    greater = float('-inf')
    for i in range(len(values)):
        if values[i] > greater:
            greater = values[i]
    return greater

values = np.random.randint(0, 100, size=15, dtype=int)
print(values)
print(greater(* values))
```
